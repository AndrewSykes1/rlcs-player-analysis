```python
from ballchaser.client import BallChaser
import config
import os
import time
import requests
```

```python
ball_chaser = BallChaser(config.api_key)
API_KEY = config.api_key
BASE_URL = 'https://ballchasing.com/api'
```

```python
# Create a list containing the group id of every group in rlcs 2024
# Check each id in said list to see if it has directly downloadable replays
# Download each replay into a folder
```

```python
# Set headers for API requests
headers = {
    'Authorization': API_KEY,
}


# Function to get group details
def get_group_details(group_id, player_name, player_id):
    url = f'{BASE_URL}/replays?player-name={player_name}&player-id=steam:{player_id}&group={group_id}'
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print(f'Successfully fetched details for group ')
        return response.json()
    else:
        print(f'Failed to get details for group : {response.status_code}')
        print(f'Response content: {response.content}')
    return None
```

```python
group_id_list = ['rlcs-2024-jsvrszynst']
final_subgroup_id_list = []
ticker = 0
save_path = r'D:\all_programming\rlcs-player-analysis\rlcs_2024_replays'

for id in group_id_list:
    ticker +=1

    try:
        group = list(ball_chaser.list_groups(group_id = id))
        time.sleep(2)
    except:
        break
    
    if len(group) != 0:
        for item in group:
            group_id_list.append(item['id'])
            if item['direct_replays'] > 0:
                final_subgroup_id_list.append(item['id'])

    print(len(group_id_list), ticker)
print(group_id_list)
print(final_subgroup_id_list)
print(id)
```
