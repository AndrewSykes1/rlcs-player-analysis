```python
import requests
import os
import config

#Player list           zen              snowyy
wanted_player = ['76561198144145654','76561198409065326']
player_order = []
total_stats = {}
deviation_stats = {}

# Replace these with your actual API key and the main group ID
API_KEY = config.api_key
MAIN_GROUP_ID = '1-major-ry36ymd607'
BASE_URL = 'https://ballchasing.com/api'

# Set headers for API requests
headers = {
    'Authorization': API_KEY,
}

# Function to get group details
def get_group_details(group_id):
    url = f'{BASE_URL}/groups/{group_id}'
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        print(f'Successfully fetched details for group {group_id}')
        return response.json()
    else:
        print(f'Failed to get details for group {group_id}: {response.status_code}')
        print(f'Response content: {response.content}')
    return None

# Main script
def main():
    # Get details for the main group
    group_details = get_group_details(MAIN_GROUP_ID)

    # Create dict structure
    for player in group_details['players']:
        if player['id'] in wanted_player:
            player_order.append(player['name'])
            for key in player['game_average']:
                total_stats[key] = {}
                for sub_key in player['game_average'][key]:
                    total_stats[key][sub_key] = []

    # Append player data to dict structure
    for player in group_details['players']:
        if player['id'] in wanted_player:
            for key in player['game_average']:
                for sub_key in player['game_average'][key]:
                    total_stats[key][sub_key].append(player['game_average'][key][sub_key])
    
    # Create dict structure for deviations
    for key in total_stats:
        deviation_stats[key] = {}
        for sub_key in total_stats[key]:
            
            lesser = min(total_stats[key][sub_key])
            greater = max(total_stats[key][sub_key])
            
            if greater != 0:
                deviation_stats[key][sub_key] = lesser / greater


    # Check deviation dict structure
    for key in deviation_stats:
        most_similar_key = max(deviation_stats[key], key=deviation_stats[key].get)
        least_similar_key = min(deviation_stats[key], key=deviation_stats[key].get)

        print(key)
        print(f'Most similar | {most_similar_key}: {round(deviation_stats[key][most_similar_key],4)}')
        print(f'Least similar | {least_similar_key}: {round(deviation_stats[key][least_similar_key],4)}')

    



    for key in total_stats:
        print(total_stats[key].keys())



#Shots: 3, 8
#Demos: 1, 2

#core
#boost
#movement
#positioning
#demo

if __name__ == '__main__':
    main()
```
